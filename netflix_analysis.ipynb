# Netflix Film Engagement Analysis

This notebook explores relationships between **popularity**, **budget**, **vote average**, and a synthetic **success** metric.

> Replace `data/netflix_movies.csv` with your real dataset when you're ready.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/netflix_movies.csv')
df.head()
```

```python
# Popularity vs Success
plt.figure()
plt.scatter(df['popularity'], df['success'])
plt.xlabel('Popularity')
plt.ylabel('Success')
plt.title('Popularity vs Success')
plt.tight_layout()
plt.show()
```

## Interpretation (student tone)

This plot helps show why **popularity** matters in our project. As popularity increases, we tend to see higher values of success.
We also built other visuals (Budget vs Success, Vote Average vs Success, Popularity vs Vote Average, and Budget vs Popularity) but this one was the clearest and easiest to explain in a presentation.
